```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd

game = pd.read_csv('games.csv')

print(game.head())
```

```python
fig, ax = plt.subplots()

ax.scatter(game['gold'], game['year'])

ax.set_title('Most Gold Medals Won')
ax.set_xlabel('Gold')
ax.set_ylabel('Year')
```

```python
from pandas.plotting import scatter_matrix

fig, ax = plt.subplots(figsize=(12,12))
scatter_matrix(game, alpha=1, ax=ax)
```
